```python
import string
import math
import re
import numpy as np
import glob
import os
from collections import Counter
from pypdf import PdfReader

pdf_folder = "Document_Corpus"
corpus = []
metadata = []
remove_table = str.maketrans("","",string.punctuation)

pdf_files = glob.glob(os.path.join(pdf_folder,"*.pdf"))

for pdf_path in pdf_files:
    filename = os.path.basename(pdf_path)

    try:
        reader = PdfReader(pdf_path)
        for page_num,page in enumerate(reader.pages):
            text = page.extract_text()
            if text and text.strip():
                cleaned_text = text.lower().translate(remove_table)

                corpus.append(cleaned_text)
                metadata.append({
                    "filename":filename,
                    "page":page_num+1,
                    "raw_text":text.strip()
                })
    
    except Exception as e:
        print(f"Error reading{filename}:{e}")
print(f"Loaded{len(corpus)} pages from {len(pdf_files)} PDF Files")
```

```python
vocab_set = set()
for sentence in corpus:
    words = sentence.split()
    vocab_set.update(words)

vocabulary = sorted(list(vocab_set))
```

```python
def compute_tf(corpus, vocabulary):
    tf_mat = np.zeros((len(corpus), len(vocabulary)), dtype=float)
    vocab_index = {term: idx for idx, term in enumerate(vocabulary)}

    for doc_id, sentence in enumerate(corpus):
        words = sentence.lower().split()
        total = len(words)
        word_counts = Counter(words)

        for word, count in word_counts.items():
            if word in vocab_index:
                col = vocab_index[word]
                tf_mat[doc_id, col] = count / total

    return tf_mat


tf_matrix = compute_tf(corpus, vocabulary)
tf_matrix.shape
```

```python
def compute_idf(corpus,vocabulary):
    idf_val = np.zeros(len(vocabulary),dtype=float)

    sentence_sets = [set(doc_sent.lower().split()) for doc_sent in corpus]
    for idx,term in enumerate(vocabulary):
        sentence_contains = sum(1 for sentence_set in sentence_sets if term in sentence_set)
        idf_val[idx] = np.log((1+len(corpus))/(1+sentence_contains))+1
    return idf_val

idf_vector = compute_idf(corpus,vocabulary)
print("IDF Vector Shape:", idf_vector.shape)
print("Example IDF values (first 5):", idf_vector[:5])
```

```python
raw_tf_idf = tf_matrix * idf_vector

row_norms = np.linalg.norm(raw_tf_idf, axis=1, keepdims=True)
row_norms[row_norms == 0] = 1.0

tf_idf_matrix = raw_tf_idf / row_norms
print("TF-IDF Matrix Shape:", tf_idf_matrix.shape)
print("Verify normalization (should be close to 1.0):", np.linalg.norm(tf_idf_matrix[0]))
```

```python
def search(query, corpus, vocabulary, tf_idf_matrix, idf_vector, top_n=5):
    remove_table = str.maketrans("","",string.punctuation)
    clean_query = query.translate(remove_table)
    query_words = clean_query.lower().split()
    total_words = len(query_words)

    if total_words == 0:
        print("Enter a Valid Query")
        return []

    query_tf = np.zeros(len(vocabulary), dtype=float)
    vocab_index = {term: idx for idx, term in enumerate(vocabulary)}
    query_words_counts = Counter(query_words)

    for word, count in query_words_counts.items():
        if word in vocab_index:
            col = vocab_index[word]
            query_tf[col] = count / total_words

    query_tf_idf = query_tf * idf_vector
    query_norms = np.linalg.norm(query_tf_idf)

    if query_norms == 0:
        print(f"NO matches found. All words in query'{query}'are out of vocabulary")
        return []
    
    query_tf_idf_norm = query_tf_idf/query_norms
    similarity_scores = np.dot(tf_idf_matrix,query_tf_idf_norm)
    top_idx = np.argsort(similarity_scores)[::-1][:top_n]

    match_found=False
    for rank,idx in enumerate(top_idx):
        score = similarity_scores[idx]
        if score>0:
            match_found=True
            meta = metadata[idx]
            print(f"Rank {rank+1} - Score {score:.2f}")
            print(f"File: {meta['filename']} (Page{meta['page']})")

            snippet = meta['raw_text'].replace('\n',' ')[:200]
            print(f"Snippet: {snippet}...")
    
    if not match_found:
        print(f"No matching document with score >0")
    
    return top_idx
```

```python
results = search("PM", corpus, vocabulary, tf_idf_matrix, idf_vector)
```
